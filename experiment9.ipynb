```python
import numpy as np

# Define two sample matrices
matrix_a = np.array([[1, 2],
                     [3, 4]])

matrix_b = np.array([[5, 6],
                     [7, 8]])

print("Original Matrix A:\n", matrix_a)
print("Original Matrix B:\n", matrix_b)

# Transpose of Matrix A
transpose_a = matrix_a.T
print("\nTranspose of Matrix A:\n", transpose_a)

# Determinant of Matrix A
determinant_a = np.linalg.det(matrix_a)
print("\nDeterminant of Matrix A:", determinant_a)

# Matrix Multiplication (A @ B)
matrix_product = matrix_a @ matrix_b
print("\nMatrix Multiplication (A @ B):\n", matrix_product)
```
